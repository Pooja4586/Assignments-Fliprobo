Question-1:

Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views


```python
#Importing required files

import selenium
import warnings
import pandas as pd
import time
import re
from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

s=Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=s)
driver.maximize_window()

#linking Site

driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")

Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]


#Getting and lisiting rank of video

try:
    ranks=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in ranks:
        Rank.append(i.text.replace(".",""))
    
except NoSuchElementException:
    Rank.append("-")

#Getting and lisiting name of video

try:
    names=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
    for i in names:
        x = re.findall("\D", i.text)
        s="".join(x)
        n=s.replace("[]","")
        Name.append(n)
    
except NoSuchElementException:
    Name.append("-")

#Getting and lisiting artist of video
    
try:
    artists=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
    for i in artists:
        Artist.append(i.text)
    
except NoSuchElementException:
    Artist.append("-")

#Getting and lisiting views of video
    
try:
    views=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
    for i in views:
        Views.append(i.text)
    
except NoSuchElementException:
    Views.append("-")
    
#Getting and lisiting upload date of video
    
try:
    upload=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
    for i in upload:
        Upload_date.append(i.text)
    
except NoSuchElementException:
    Upload_date.append("-")
    
df=pd.DataFrame({"Rank":Rank,"Name of Video":Name,"Artist":Artist,"Total Views(In Billions)":Views,"Date of Upload":Upload_date})
df
```

Question-5:

Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
#Importing required files

import selenium
import warnings
import pandas as pd
import time
from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

s=Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=s)
driver.maximize_window()

#linking Site

driver.get("https://github.com")

time.sleep(10)
try:
    explore=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div[3]/nav/a[5]")
    explore.click()

    time.sleep(3)

    trending=driver.find_element(By.XPATH,"/html/body/div[1]/div[5]/main/div[1]/nav/div/a[3]")
    trending.click()

    time.sleep(3)
except:
    driver.get("https://github.com/trending")

Repository_title=[]
Repository_description=[]
Contributors=[]
Language_used=[]
Link=[]

#Getting repository links
try:
    url=driver.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]/a')
    for i in url :
        link=i.get_attribute("href")
        Link.append(link)
        
except NoSuchElementException:
    Link.append("-")

#Getting and lisiting name of repository

try:
    name=driver.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]')
    for i in name:
        Repository_title.append(i.text)
        
except NoSuchElementException:
    Repository_title.append("-")

#Getting and lisiting description of repository

try:
    desc=driver.find_elements(By.XPATH,'//p [@class="col-9 color-fg-muted my-1 pr-4"]')
    for i in desc:
        Repository_description.append(i.text)
        
except NoSuchElementException:
    Repository_description.append("-")
    

#Getting and lisiting language

try:
    language=driver.find_elements(By.XPATH,'//span [@class="d-inline-block ml-0 mr-3"]')
    for i in language:
        Language_used.append(i.text)
except NoSuchElementException:
    Language_used.append("-")
    

#Getting and lisiting contributors    

n=len(Link)

for i in range(0,n):
    
    contri=Link[i]
    driver.get(contri)
    time.sleep(3)
    try:
        count=driver.find_element(By.XPATH,'//h2[@class="h4 mb-3"]/a[contains(text(),"Contributors")]/span')
        Contributors.append(count.text)
    except NoSuchElementException:
        Contributors.append("-")

    Lang=[]
    try:
        lang=driver.find_elements(By.XPATH,"//span[@class='color-fg-default text-bold mr-1']")
        
        if lang:
            for j in lang:
                Lang.append(j.text)
        else:
            Lang.append('-')

        Language_used.append(Lang)

    except NoSuchElementException:
        Language_used.append('-')
        
df=pd.DataFrame({"Repository Title":Repository_title,"Repository Description":Repository_description,
                 "Language Used ":Language_used[23:],"Number of Contributors":Contributors})
                        
df
```

```python
import selenium
import warnings
import pandas as pd
import time
from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

s=Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=s)
driver.maximize_window()

#linking Site

driver.get("https://archive.ics.uci.edu/ml/index.php")

time.sleep(5)


alldata=driver.find_element(By.XPATH,'/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b')
alldata.click()

time.sleep(3)

Name=[]
Datatype=[]
Task=[]
Attribute=[]
Instance=[]
num_att=[]
Year=[]

try:
    name=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")

    for i in name:
        Name.append(i.text)     #Scraping name list

except NoSuchElementException:
    Name.append("-")

try:
    data=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p')
    for i in data:
        Datatype.append(i.text)          #Scraping datatype

except NoSuchElementException:
    Datatype.append("-")

try:
    task=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p')

    for i in task:
        Task.append(i.text)     #Scraping default task

except NoSuchElementException:
    Task.append("-")
    
try:
    attrib=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p')
    for i in attrib:
        Attribute.append(i.text)      #Scraping attribute

except NoSuchElementException:
    Attribute.append("-")
    
    
try:
    instance=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p')
    for i in instance:
        Instance.append(i.text)            #Scraping number of instances

except NoSuchElementException:
    Instance.append("-")
    
try:
    num=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p')
    for i in num:
        num_att.append(i.text)            #Scraping number of attributes

except NoSuchElementException:
    num_att.append("-")
    

try:
    year=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p')
    for i in year:
        Year.append(i.text)            #Scraping Year

except NoSuchElementException:
    Year.append("-")    

df=pd.DataFrame({"Name":Name,"Data Type":Datatype[1:],"Default Task":Task[1:],
                 "Attribute":Attribute[1:],"Number of Instance":Instance[1:],
                 "Number of Attributes":num_att[1:],"Year":Year[1:]})

df
```
